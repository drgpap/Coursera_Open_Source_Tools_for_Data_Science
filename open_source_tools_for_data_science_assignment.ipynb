### The code is generating a matrix of random entries and calculate LU decomposition and visualizes the upper and lower triangular matrices


```python
import numpy as np
```

```python
import scipy.linalg as scl
```

```python
import matplotlib.pyplot as mpl
```

```python
ns = 200
```

```python
Ar = np.random.randn(ns,ns)
```

```python
P, L, U = scl.lu(Ar)
```

```python
mpl.pcolor(L)
```

```python
mpl.pcolor(U)
```

***
Above code has the following steps:
1. Initialize a parameter "ns" with size of square matrix
2. Generate a matrix "Ar" of size ns filled with random entries
3. Perform an LU decomposition of the matrix returning three matrics:
    - L: Lower triangular matrix
    - U: Upper triangular matrix
    - P: Permutation matrix
4. Vizualise L matrix
5. Vizualise U matrix
